## Density Based Clustering

### Import

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

### Example 1 (Make Moons)

```python
x , _ = make_moons(n_samples= 500 , noise=0.085 , random_state=8 )
x.shape
plt.scatter(x[:,0] , x[:,1])
```

```python
dbscan = DBSCAN(eps=0.2 , min_samples=5)
dbscan_predict = dbscan.fit_predict(x)

plt.scatter(x[:,0] , x[:,1] , c= dbscan_predict)
```

### Example 2 (Face)

```python
# Face outline 
theta = np.linspace(0, 2*np.pi, 500)
x_circle = 10 * np.cos(theta) + np.random.normal(0, 0.2, theta.size)
y_circle = 10 * np.sin(theta) + np.random.normal(0, 0.2, theta.size)

# Eyes 
x_eye1 = np.random.normal(-3, 0.5, 50)
y_eye1 = np.random.normal(3, 0.5, 50)

x_eye2 = np.random.normal(3, 0.5, 50)
y_eye2 = np.random.normal(3, 0.5, 50)

# Mouth 
theta_mouth = np.linspace(0, np.pi, 200)
x_mouth = 5 * np.cos(theta_mouth) + np.random.normal(0, 0.2, theta_mouth.size)
y_mouth = -5 - 2 * np.sin(theta_mouth) + np.random.normal(0, 0.2, theta_mouth.size)

# Merging all points
X = np.vstack([
    np.column_stack([x_circle, y_circle]),
    np.column_stack([x_eye1, y_eye1]),
    np.column_stack([x_eye2, y_eye2]),
    np.column_stack([x_mouth, y_mouth])
])
```

```python
clustering = DBSCAN(eps=0.8, min_samples=5).fit(X)
labels = clustering.labels_
```

```python
plt.figure(figsize=(6,6))
plt.scatter(X[:,0], X[:,1], c=labels, cmap='rainbow', s=20)
plt.axis("equal")
plt.show()
```
